```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
S = 100
K_CALL = S*1.1
K_PUT = S*0.9
K_PUT2 = S*0.6
```

```python
sample = np.linspace(0, S*2, 100)
```

```python
payoff_collar = (sample-S) - np.clip(sample - K_CALL, 0, np.inf) + np.clip(K_PUT - sample, 0, np.inf)
plt.plot(payoff_collar)
plt.title('Collar')
plt.show()
```

```python
payoff_3way = (sample-S) - np.clip(sample - K_CALL, 0, np.inf) + np.clip(K_PUT - sample, 0, np.inf) - np.clip(K_PUT2 - sample, 0, np.inf)
plt.plot(payoff_3way)
plt.title('3-Way')
plt.show()
```

```python
K_CALL2 = S*1.2
```

```python
payoff_1by2 = -np.clip(sample - K_CALL, 0, np.inf) + 2*np.clip(sample - K_CALL2, 0, np.inf)
plt.plot(payoff_1by2)
plt.title('Call Spread Ratio')
plt.show()
```
